```python
df=pd.read_csv('http://stats.moe.gov.tw/files/detail/107/107_student.csv')
df.head()
```

```python
colname = df.loc[1]
colname[1]
#查看第一列的資料
```

```python
colname = df.loc[1].values
#colname[1]
```

```python
df = df[2:]
df.head()
```

```python
#每次都要先做一些資料清理的動作
#很重要
```

```python
df.columns = colname
df.head()
```

```python
colname
```

```python
(df['日間∕進修別']=='D 日') & (df['等級別'] == 'B 學士')
```

```python
undergraduate = df[(df['日間∕進修別']=='D 日') & (df['等級別'] == 'B 學士')]  #直接搜尋指定項目
```

```python
undergraduate.head()
```

```python
df2 = undergraduate.loc[:,'學校名稱':'女生計'].copy() #將其copy出來，loc是指地點，一開始的:是列全拿，然後逗號後面是行的範圍
```

```python
#df2.女生計 / df2.男生計 #這樣不行，因為要把逗號拿掉才行
```

```python
import locale

#atof #把他變成浮點數
```

```python
import locale
locale.setlocale(locale.LC_NUMERIC,'')    #給他一個空的字串，讓他變成台灣的UTF8
```

```python
df2['男生計']=df2['男生計'].apply(locale.atof) #老師說，也可以一個個去，但這方法更好
```

```python
df2['女生計']=df2['女生計'].apply(locale.atof)
```

```python
df2['女男比例'] = df2.女生計 / df2.男生計 #計算男女比 #並新增一行女男比例
```

```python
df2.head()
```

```python
#找最高最低者
df3 = df2.sort_values(by=['女男比例'], ascending=False) #ascending=false是從大排到小
#pandas常常把我們要的存程一個新的dataframe
```

```python
df3[:20].plot.bar(x='學校名稱',y=['女男比例']) #前20個
```
