```python
from IPython.display import clear_output
import numpy as np
import time

sp = 0
emoticon = '咕嚕靈波 （●′∀‵）ノ♡ '

for i in range(100):
    k=(-1)**np.random.randint(1,3)
    sp=sp+k
    if sp<0:
        sp=0
    elif sp >50:
        sp=50
    clear_output(wait=True)
    print(' '*sp +emoticon)
    time.sleep(0.5) #停頓0.5秒
    
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
x=[1,2,3,4,5]
y1=[1,2,3,4,5]
y2=[5,4,3,2,1]
plt.plot(x,y1, label = 'y1') #可以協助顯示誰是誰
plt.plot(x,y2, label = 'y2') 

plt.legend(loc ='best') #顯示位置

plt.xlabel("X")
plt.ylabel("Y")

plt.xlim(-2,20) #X軸顯示範圍
plt.ylim(-20,10) #Y軸顯示範圍

plt.xticks(np.arange(-2,20,5)) #讓他每五格一個區段
plt.xticks(np.arange(-2,20,5))

plt.title("Title")
```

```python
import seaborn as sns
sns.set_style("whitegrid") #白底 #seaborn自動標網格
plt.plot(x,y1)
plt.plot(x,y2)
tips = sns.load_dataset('tips') #SNS有內建資料庫可以讀取
```
